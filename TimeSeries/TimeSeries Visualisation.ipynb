```python
import pandas as pd
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import datetime
```

---
# Visualising TimeSeries
## first look

```python
N = 360
df = pd.DataFrame()
## make random points
df['t'] = np.random.randint(low=1, high=N, size=N)
## calculate sin
df['y'] = np.sin(df.t * np.pi / 180)
## sorting dataframe
df1 = df.sort_values(by='t').copy()

## vis unsorted data
f, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].plot(df.t, df.y)
ax[0].grid()

## vis sorted data
ax[1].plot(df1.t, df1.y)
ax[1].grid()
plt.show()
```

### You mast be sure, you data sorting before you try to visualise it !...
But you can use method **scatter** instead **plot** to plot scatter graph


```python
f, ax = plt.subplots(figsize=(5,5))
ax.scatter(df.t, df.y)
ax.grid()
plt.show()
```

## visualising x-scale as time
Hallmark of time series visualization is x-scale as time, not integer or object
*make visualising with pandas*

```python
df = pd.DataFrame()
## make 365 days from 2019-09-01
df['t'] = pd.date_range(start='2019-09-01', periods=365, freq='D')
## calculate sin from index of days
df['y'] = np.sin(df.index * np.pi / 180)
```

```python
## make vis with pandas
## plot line
df.plot('t','y')
df.head(2)
```

```python
## bar plot
df.plot(x='t', y='y', kind='bar')
df.head(2)
```
